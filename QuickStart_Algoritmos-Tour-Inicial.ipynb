# Quick Start para Algoritmos de Tour Inicial
Esta es una guía para ejecutar los algoritmos que dan un Tour Inicial para la solución de TSP.

### Importación de Módulos
Primero se importan los módulos necesarios para la ejecución de este programa.

```python
import sys
sys.path.append('tsp')
import os
from os import path, listdir
from tsp import benchmarks,metrics,tspplot
from tsp.tspobject import TSPObject
from tsp import initialtour

benchmark_path = r'benchmarks'
```

### Selección de Benchmark
En esta sección, se especifica el archivo .tsp correspondiente a un benchmark de TSP.

```python
test_file = path.join(benchmark_path,'dj38.tsp')
```

### Creación de Objeto TSP
Después se crea un objeto de la clase TSPObject en base al benchmark seleccionado anteriormente.
También se llama su función M(), que genera una matriz de distancias, necesaria para la ejecución de los algoritmos.

```python
O = TSPObject(test_file,metrics.euc_dist)
print(O.M())
```

## Tour Inicial Aleatorio
Para obtener un Tour Inicial Aleatorio se llama la función 'random_tour', enviando como parámetro el objeto TSPObject.

```python
tour_inicial = initialtour.random_tour(O)
```

A continuación se grafica el Tour Inicial obtenido:

```python
tspplot.plot_tsp(O,tour_inicial,figsize=(6,6))
```

## Algoritmo Christofides
Para obtener un Tour Inicial utilizando el Algoritmo Christofides se llama la función 'tsp_c', enviando como parámetro el objeto TSPObject.

```python
tour_inicial = initialtour.tsp_c(O)
```

A continuación se grafica el Tour Inicial obtenido:

```python
tspplot.plot_tsp(O,tour_inicial,figsize=(6,6))
```

## Algoritmo Greedy
Para obtener un Tour Inicial utilizando el Algoritmo Greedy se llama la función 'TSP_greedy', enviando como parámetro el objeto TSPObject y el índice del nodo inicial. En este caso el índice será 0.

```python
tour_inicial = initialtour.TSP_greedy(O,0)
```

A continuación se grafica el Tour Inicial obtenido:

```python
tspplot.plot_tsp(O,tour_inicial,figsize=(6,6))
```

## Algoritmo Best-Greedy
Para obtener un Tour Inicial utilizando el Algoritmo Best-Greedy se llama la función 'Best_TSP_greedy', enviando como parámetro el objeto TSPObject.

```python
tour_inicial = initialtour.Best_TSP_greedy(O)
```

A continuación se grafica el Tour Inicial obtenido:

```python
tspplot.plot_tsp(O,tour_inicial,figsize=(6,6))
```
