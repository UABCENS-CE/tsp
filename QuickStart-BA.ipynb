# Quick Start for Bat Algorithm

En esta sección se importan los módulos necesarios para ejecutar el programa.

```python
import sys
sys.path.append('tsp')
import os
from os import path, listdir
from tsp import benchmarks,metrics,tspplot
from tsp import BA
from tsp.tspobject import TSPObject

benchmark_path = r'benchmarks'
```

# Selección de Benchmark

Mediante esta linea de código se selecciona el benchmark con el que se ejecutará el código.

```python
tsp_data = path.join(benchmark_path,'eil51.tsp')
```

# Creación de TSPObject

En esta sección se crea un objeto TSPObject; este objeto guarda todos los datos del benchmark. Tambíen se crea la matriz de distancias, la cual se encarga de crear un diccionario con todas las distancias de cada nodos respecto a sus vecinos.

```python
O = TSPObject(tsp_data,metrics.euc_dist)
O.M()
```

# Implementación del algoritmo

En esta linea de código se implementa el algoritmo. La función necesita como parámetros el objeto TSPObject, la cantidad de iteraciones y el tamaño de la población

```python
iteraciones = 100
tam_poblacion = 15
tour = BA.bat_algorithm(O, iteraciones, tam_poblacion)
```

# Graficar tour resultante

Aqui se grafica la solución determinada por el algoritmo.

```python
print("****RESULTADO****")
tspplot.plot_tsp(O,tour=tour,figsize=(6,6))
```
