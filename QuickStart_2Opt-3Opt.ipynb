# Quick Start para Algoritmos 2-Opt y 3-Opt
Esta es una guía para ejecutar los algoritmos 2-Opt y 3-Opt para la solución de TSP.

### Importación de Módulos
Primero se importan los módulos necesarios para la ejecución de este programa.

```python
import sys
sys.path.append('tsp')
import os
from os import path, listdir
from tsp import benchmarks,metrics,tspplot
from tsp.tspobject import TSPObject
from tsp import initialtour
from tsp import tsp3opt
from tsp import tsp2opt
import numpy as np

benchmark_path = r'benchmarks'
```

### Selección de Benchmark
En esta sección, se especifica el archivo .tsp correspondiente a un benchmark de TSP.

```python
test_file = path.join(benchmark_path,'dj38.tsp')
```

### Creación de Objeto TSP
Después se crea un objeto de la clase TSPObject en base al benchmark seleccionado anteriormente.
También se llama su función M(), que genera una matriz de distancias, necesaria para la ejecución de los algoritmos.

```python
O = TSPObject(test_file,metrics.euc_dist)
print(O.M())
```

## Algoritmo 2-Opt
Para obtener la soloción al TSP con el Algoritmo 2-Opt se llama la función 'tsp_2opt', enviando como parámetro el objeto TSPObject y un Tour Inicial.

El Tour Inicial se puede generar con alguno de los algoritmos incluidos en este programa.
En este caso se el Tour Inicial se generará con el Algoritmo Greedy, enviando un índice aleatorio.

```python
# Indice Aleatorio
i = np.random.randint(len(O.V))

# Obtener Tour Inicial
tour_inicial = initialtour.TSP_greedy(O,i)

# Ejecución de ALgoritmo 2-Opt
tour_final = tsp2opt.tsp_2opt(O,tour_inicial)
```

Gráfica del Tour Inicial:

```python
tspplot.plot_tsp(O,tour_inicial,figsize=(6,6))
```

Gráfica del Tour Final:

```python
tspplot.plot_tsp(O,tour_final,figsize=(6,6))
```
