## Divide and Conquer
La implementacion del algoritmo esta dividida en 2 partes:
La primera parte se encarga de dividir el mapa en 2 regiones y regresa 3 variables que son necesarias para la segunda parte (la region 1 y 2 y la ciudad comun entre ellos). La segunda parte se encarga de unir los tours en uno solo.

### Primera parte
Esta primera parte crea 2 archivos: reg1.tsp y reg2.tsp.
Estos 2 archivos contienen las ciudades correspondientes para cada region y el tour debe resolverse.

```python
variables = DAC.tsp_dac(O)
```

### Segunda parte
Para esta segunda parte los archivos reg1.tsp y reg2.tsp deben de estar resueltos y grabados en un archivo .cyc (Que es el formato en el cual los guarda concorde).
Las variables que son obtenidas de la primera parte se envian como parametro en esta segunda parte.

```python
tour = DAC.dac_tour(O,variables)
```
